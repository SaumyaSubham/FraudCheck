First of all, Welcome Everyone!

This is a project of Machine Learning named as FraudCheck which is basically a online fraud detection system. The ML is implemented by using DecisionTreeClassifier from sklearn.tree module.

The data are preprocessed and distributed by pie chart with the help of plotly module.

The steps are must be followed:
1. Check if there are null value.
2. Explore the Transactions type.
3. Check Correlation.
4. Split the data.
5. Train the ML model.

So, Let's Start!!

```python
import pandas as pd
import numpy as np
data = pd.read_csv('onlinefraud.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Exploring transaction types
data.type.value_counts()
```

```python
type = data['type'].value_counts()
transactions = type.index
quantity = type.values

import plotly.express as px
figure = px.pie(data,
                values = quantity,
                names = transactions, hole = 0.5,
                title = 'Distribution of Transaction type')
figure.show()
```

```python
# Checking correlation
correlation = data.corr()
print(correlation["isFraud"].sort_values(ascending = False))
```

```python
data['type'] = data['type'].map({'CASH_OUT' : 1, 'PAYMENT' : 2,
                                 'CASH_IN' : 3, 'TRANSFER' : 4,
                                 'DEBIT' : 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
```

```python
data.head()
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
data.dropna(inplace=True)
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
# Training the ML model
from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.10, random_state=42)
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Prediction
# features = [type, amount, oldbalanceOrg, newbalanceOrig]
features = np.array([[4, 9000.60, 9000.60, 0.0]])
print(model.predict(features))
```
